## Drowsiness Detection for Human Safety

### Overview
*Kecelakaan lalu lintas masih menjadi isu global yang sangat serius. Menurut laporan World Health Organization (2023), lebih dari 1,19 juta orang meninggal setiap tahun akibat kecelakaan di jalan raya. Kondisi kelelahan dan drowsiness pada pengemudi merupakan salah satu faktor utama yang berkontribusi terhadap tingginya angka tersebut, karena menurunkan kewaspadaan, memperlambat respons, dan meningkatkan risiko kehilangan kontrol saat berkendara.*

*Seiring perkembangan teknologi, pendekatan berbasis machine learning dan computer vision semakin banyak dimanfaatkan dalam pengembangan Driver Monitoring Systems. Berbagai penelitian menunjukkan bahwa sistem otomatis memiliki potensi besar dalam membantu mengidentifikasi kondisi drowsiness pada pengemudi. Meskipun begitu, pengembangan model drowsiness detection tetap menghadapi berbagai tantangan, mulai dari kondisi lingkungan yang beragam, perbedaan karakteristik pengemudi, hingga kebutuhan sistem yang mampu bekerja secara stabil dan cepat dalam situasi berkendara yang berbeda-beda.*

### Aim

*Dengan demikian, eksperimen ini bertujuan untuk mengembangkan model berbasis computer vision yang mampu menghitung jumlah kerumunan secara akurat dari gambar. Model ini akan mendukung pengambilan keputusan secara real-time untuk mencegah kemacetan, mengoptimalkan evakuasi, dan memastikan pengalaman pengunjung yang aman dan menyenangkan selama Festival Harmoni Nusantara.*

## Konversi Video menjadi *sequence* gambar

````python
def extract_frames_from_videos(train_dir):
    train_dir = Path(train_dir)

    category_map = {'drowsiness': 1, 'non-drowsiness': 0}

    for label in [0, 1]:
        (train_dir / str(label)).mkdir(parents=True, exist_ok=True)

    video_files = []
    for category in ['drowsiness', 'non-drowsiness']:
        category_path = train_dir / category
        if not category_path.exists():
            continue

        for subject_dir in sorted(category_path.iterdir()):
            if not subject_dir.is_dir():
                continue

            for video_file in subject_dir.glob('*.mp4'):
                video_files.append((video_file, category, subject_dir.name))

    for video_path, category, subject in tqdm(video_files, desc="Processing videos"):
        label = category_map[category]
        video_name = video_path.stem

        cap = cv2.VideoCapture(str(video_path))

        sequence_idx = 1
        frames_buffer = []

        while True:
            ret, frame = cap.read()
            if not ret:
                break

            frames_buffer.append(frame)

            if len(frames_buffer) == 30:
                seq_name = f"{subject}-{video_name}-seq_{sequence_idx:05d}"
                seq_dir = train_dir / str(label) / seq_name
                seq_dir.mkdir(parents=True, exist_ok=True)

                for i, img in enumerate(frames_buffer, start=1):
                    filename = f"{i:02d}.jpg"
                    output_path = seq_dir / filename
                    cv2.imwrite(str(output_path), img)

                frames_buffer = []
                sequence_idx += 1

        cap.release()
    
if not (Settings.TRAIN_DIR / '0').exists() or not (Settings.TRAIN_DIR / '1').exists():
    extract_frames_from_videos(Settings.TRAIN_DIR)
else:
    print("Processing videos: 100%|██████████| 120/120 [45:33<00:00, 22.78s/it]")
````

## Loading ke DataFrame

````python
def load_data(train_dir, test_dir):
    train_dir = Path(train_dir)
    test_dir = Path(test_dir)

    train_data = []
    for label_folder in ['0', '1']:
        label_path = train_dir / label_folder
        if not label_path.is_dir():
            continue

        label = int(label_folder)

        for seq_dir in label_path.iterdir():
            if not seq_dir.is_dir():
                continue

            parts = seq_dir.name.split('-')
            subject = parts[0]
            video_type = f"{parts[1]}_{parts[2]}" if len(parts) >= 3 else "unknown"

            train_data.append({
                'subject': subject,
                'video_type': video_type,
                'seq_path': str(seq_dir),
                'label': label
            })

    test_data = []
    for seq_dir in test_dir.iterdir():
        if seq_dir.is_dir():
            test_data.append({
                'seq_id': seq_dir.name,
                'seq_path': str(seq_dir)
            })

    train_df = pd.DataFrame(train_data)
    test_df = pd.DataFrame(test_data)

    return train_df, test_df
````

## Image Dataset Loader

````python
class SequenceDataset(Dataset):
    def __init__(self, df, transform=None, has_labels=True):
        self.df = df.reset_index(drop=True)
        self.transform = transform
        self.has_labels = has_labels

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        seq_path = Path(self.df.iloc[idx]['seq_path'])

        frames = []
        for i in range(1, 31):
            frame_path = seq_path / f"{i:02d}.jpg"
            if frame_path.exists():
                image = Image.open(frame_path).convert("RGB")
                if self.transform:
                    image = self.transform(image)
                frames.append(image)

        frames = torch.stack(frames)

        if self.has_labels:
            label = torch.tensor(self.df.iloc[idx]['label'], dtype=torch.long)
            return frames, label
        else:
            return frames, self.df.iloc[idx]['seq_id']

def collate_sequences(batch):
    all_images = []
    all_labels = []

    for images, label in batch:
        for img in images:
            all_images.append(img)
            all_labels.append(label)

    return torch.stack(all_images), torch.stack(all_labels)
````

````python
def load_sequences_by_subject(train_df):
    subject_sequences = {}

    for _, row in train_df.iterrows():
        subject = row['subject']
        if subject not in subject_sequences:
            subject_sequences[subject] = []
        subject_sequences[subject].append((Path(row['seq_path']), row['label']))

    return subject_sequences
````

````python
subject_sequences = load_sequences_by_subject(train_df)
subjects = list(subject_sequences.keys())
random.shuffle(subjects)

train_size = int(0.8 * len(subjects))
train_subjects = subjects[:train_size]
val_subjects = subjects[train_size:]

train_subset_df = train_df[train_df['subject'].isin(train_subjects)].reset_index(drop=True)
val_subset_df = train_df[train_df['subject'].isin(val_subjects)].reset_index(drop=True)

train_dataset = SequenceDataset(train_subset_df, transform=transform, has_labels=True)
val_dataset = SequenceDataset(val_subset_df, transform=transform, has_labels=True)

train_loader = DataLoader(train_dataset, batch_size=4, shuffle=True, collate_fn=collate_sequences)
val_loader = DataLoader(val_dataset, batch_size=4, shuffle=False, collate_fn=collate_sequences)

print(f"Train subjects: {len(train_subjects)}, Val subjects: {len(val_subjects)}")
print(f"Train sequences: {len(train_subset_df)}, Val sequences: {len(val_subset_df)}")

for images, labels in train_loader:
    print("Batch images shape:", images.shape)
    print("Batch labels:", labels)
    break
````

## Model Loader

````python
class MobileNetV2(nn.Module):
    def __init__(self, pretrained=True):
        """
        Model using MobileNetV2.

        Args:
            pretrained (bool): If True, load pre-trained from ImageNet.
            dropout_prob (float): Dropout probability for regularization.
        """
        
        super(MobileNetV2, self).__init__()
        
        # Load pre-trained MobileNetV2
        self.mobilenet = mobilenet_v2(pretrained=pretrained)
        
        # Take input dimension from fully connected (fc) layer
        in_features = self.mobilenet.classifier[1].in_features
        
        # Replace the classifier with binary classification
        self.mobilenet.classifier = nn.Sequential(
            nn.Linear(in_features, 1)                           # Output 1 logit for binary classification
        )
    
    def forward(self, pixel_values):
        """
        Forward pass for input processing through MobileNetV2.

        Args:
            pixel_values (torch.Tensor): Input tensor with shape (batch_size, 3, 224, 224).

        Returns:
            logits (torch.Tensor): Logit values for binary classification.
        """
        return self.mobilenet(pixel_values)
````

````python
def evaluate_model(model, dataloader, device, loss_function, threshold_default=0.5):
    model.eval()
    total_samples = 0
    correct_predictions = 0
    all_true_labels = []
    all_predicted_probs = []
    total_val_loss = 0.0

    # Use torch.no_grad() to disable gradient computation
    with torch.no_grad():

        # Iterate over the data
        for batch_idx, (pixel_values, labels) in enumerate(dataloader):
            pixel_values = pixel_values.to(device)
            labels = labels.to(device).float()

            # Forward pass
            logits = model(pixel_values).squeeze(1)  
            probs = torch.sigmoid(logits)

            # Compute loss
            val_loss = loss_function(logits, labels)
            total_val_loss += val_loss.item()

            # Store the predicted probabilities and true labels
            all_predicted_probs.extend(probs.cpu().numpy())
            all_true_labels.extend(labels.cpu().numpy())

            # Final predictions
            preds = (probs > threshold_default).long()

            # Compute the number of correct predictions
            correct_predictions += (preds == labels).sum().item()

            # Total samples
            total_samples += labels.size(0)

    # Average validation loss
    avg_val_loss = total_val_loss / len(dataloader)

    # Accuracy
    accuracy = correct_predictions / total_samples * 100

    # Precision-Recall Curve
    precision, recall, _ = precision_recall_curve(all_true_labels, all_predicted_probs)

    # F1 Scores
    f1_scores = 2 * (precision * recall) / (precision + recall + 1e-7)

    # Evaluate the model
    preds_default = (np.array(all_predicted_probs) > threshold_default).astype(int)
    precision_default = precision_score(all_true_labels, preds_default)
    recall_default = recall_score(all_true_labels, preds_default)
    f1_default = f1_score(all_true_labels, preds_default)

    # Print the validation metrics
    print(f"Validation Metrics:")
    print(f"Average Validation Loss: {avg_val_loss:.6f}")
    print(f"Accuracy: {accuracy:.2f}%")
    print(f"Precision: {precision_default:.2f}")
    print(f"Recall: {recall_default:.2f}")
    print(f"F1 Score: {f1_default:.2f}")
    print()

    # Return metrics dan average validation loss
    return {
        "accuracy": accuracy,
        "avg_val_loss": avg_val_loss,
        "precision_default": precision_default,
        "recall_default": recall_default,
        "f1_default": f1_default,
        "precision": precision,
        "recall": recall,
        "f1_scores": f1_scores,
    }
````

````python
print("Training started...")
print("-" * 75)

# Training loop
# Iterate over epochs
for epoch in range(Settings.EPOCHS):

    # Set the model to training mode
    model.train()

    # Initialize total loss for the epoch
    total_loss_epoch = 0

    # Initialize lists to store predictions and labels
    all_train_preds = []
    all_train_labels = []

    # Iterate over the training data
    for pixel_values, labels in tqdm(train_loader, desc=f"Epoch [{epoch+1}/{Settings.EPOCHS}]"):

        # Move data to device
        pixel_values, labels = pixel_values.to(device), labels.to(device).float()

        # Zero the gradients
        optimizer.zero_grad()

        # Forward pass
        logits = model(pixel_values).squeeze(1)     
        loss = lossFunc(logits, labels)            

        # Backward pass 
        loss.backward()

        # Clip the gradients
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)

        # Update the parameters
        optimizer.step()

        # Store the loss
        total_loss_epoch += loss.item()

        # Keep track of all predictions and labels for training metrics
        probs = torch.sigmoid(logits)
        preds = (probs > 0.5).long().cpu().numpy()
        all_train_preds.extend(preds)
        all_train_labels.extend(labels.cpu().numpy())

    # Count training metrics
    train_accuracy = np.mean(np.array(all_train_preds) == np.array(all_train_labels)) * 100
    train_precision = precision_score(all_train_labels, all_train_preds, zero_division=1)
    train_recall = recall_score(all_train_labels, all_train_preds, zero_division=1)
    train_f1 = f1_score(all_train_labels, all_train_preds, zero_division=1)

    # Keep training metrics
    avg_loss = total_loss_epoch / len(train_loader)
    train_losses.append(avg_loss)
    train_accuracies.append(train_accuracy)
    train_precisions.append(train_precision)
    train_recalls.append(train_recall)
    train_f1_scores.append(train_f1)

    # Print training metrics
    print(f"Epoch [{epoch+1}/{Settings.EPOCHS}], Average Training Loss: {avg_loss:.6f}")
    print()
    print(f"Training Metrics:")
    print(f"Accuracy: {train_accuracy:.2f}%")
    print(f"Precision: {train_precision:.2f}")
    print(f"Recall: {train_recall:.2f}")
    print(f"F1 Score: {train_f1:.2f}")
    print()

    # Evaluate model on validation set
    metrics = evaluate_model(model, val_loader, device, lossFunc)
    val_loss_epoch = metrics["avg_val_loss"]
    val_accuracy_epoch = metrics["accuracy"]
    val_precision_epoch = metrics["precision_default"]
    val_recall_epoch = metrics["recall_default"]
    val_f1_score_epoch = metrics["f1_default"]

    # Keep validation metrics
    val_losses.append(val_loss_epoch)
    val_accuracies.append(val_accuracy_epoch)
    val_precisions.append(val_precision_epoch)
    val_recalls.append(val_recall_epoch)
    val_f1_scores.append(val_f1_score_epoch)

    # Save the best model based on F1 Score
    if val_f1_score_epoch > best_f1 or (val_f1_score_epoch == best_f1 and val_loss_epoch < best_loss):
        best_f1 = val_f1_score_epoch
        best_loss = val_loss_epoch
        torch.save(model.state_dict(), Settings.MODEL_PTH)
        print(f"Best model saved with F1 Score: {best_f1:.2f}")
    
    print("-" * 75)

metrics_df = pd.DataFrame({
    "epoch": range(1, Settings.EPOCHS + 1),
    "train_loss": train_losses,
    "val_loss": val_losses,
    "train_accuracy": train_accuracies,
    "val_accuracy": val_accuracies,
    "train_precision": train_precisions,
    "val_precision": val_precisions,
    "train_recall": train_recalls,
    "val_recall": val_recalls,
    "train_f1": train_f1_scores,
    "val_f1": val_f1_scores
})

metrics_df.to_csv(Settings.METRICS_CSV, index=False)
print("Training finished.")
````

````python
def predict_and_save(model, test_df, output_csv, transform, threshold=0.5, batch_size=4):
    model.eval()
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    test_dataset = SequenceDataset(test_df, transform=transform, has_labels=False)
    test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

    sequence_predictions = {}

    with torch.no_grad():
        for frames_batch, seq_ids in tqdm(test_loader, desc="Predicting sequences"):
            batch_size_actual = frames_batch.size(0)

            for i in range(batch_size_actual):
                frames = frames_batch[i]
                seq_id = seq_ids[i]

                frames = frames.to(device)
                logits = model(frames).squeeze()
                probs = torch.sigmoid(logits)

                avg_prob = probs.mean().item()
                sequence_predictions[seq_id] = 1 if avg_prob > threshold else 0

    submission_df = pd.DataFrame({
        'id': list(sequence_predictions.keys()),
        'label': list(sequence_predictions.values())
    })

    submission_df.to_csv(output_csv, index=False)
    print(f"Predictions saved to {output_csv}")

# Optimal threshold
opt_threshold = 0.5

# Predict and save results
predict_and_save(
    model=best_model,
    test_df=test_df,
    output_csv=Settings.SUBMISSION_CSV,
    transform=transform,
    threshold=opt_threshold,
    batch_size=4
)
````
